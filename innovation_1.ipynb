```python
import gradio as gr

def generate(process, defect):
    prompt = f"close-up of metal surface after {process} showing {defect} defect, industrial inspection, no humans, high detail"
    image = pipe(prompt).images[0]
    return image

demo = gr.Interface(
    fn=generate,
    inputs=[
        gr.Dropdown(["welding", "casting","molding","forming","machining","joining","additive manufacturing","cutting","punching","forging","extrusion"], label="Select Process"),
        gr.Dropdown(["crack", "porosity","crater","spatter","lamellar tearing","distortion","misallignment","whiskers"], label="Select Defect")
    ],
    outputs="image",
    title="AI Manufacturing Defect Generator",
    description="Generate realistic defect images using AI"
)

demo.launch()
```
