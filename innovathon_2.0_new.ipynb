```python
from PIL import Image
import gradio as gr

def generate(process, defect):
    try:
        file_path = f"{process}_{defect}.png"
        return Image.open(file_path)
    except:
        return "Image not found"

demo = gr.Interface(
    fn=generate,
    inputs=[
        gr.Dropdown(["welding", "casting","molding","forming","machining","joining","additive manufacturing","cutting","punching","forging","extrusion"], value="welding", label="Process"),
        gr.Dropdown(["crack", "porosity","crater","spatter","lamellar tearing","distortion","misalignment","whiskers"], value="crack", label="Defect")
    ],
    outputs="image",
    title="AI Manufacturing Defect Generator",
    description="Generate defect images based on process and defect type"
)

demo.launch()
```

```python
import gradio as gr
from PIL import Image

def generate(process, defect):
    try:
        file_path = f"{process}_{defect}.png"
        return Image.open(file_path)
    except:
        return None

demo = gr.Interface(
    fn=generate,
    inputs=[
        gr.Dropdown(
            ["welding", "casting","machining"],
            value="welding",
            label="Select Process"
        ),
        gr.Dropdown(
            ["crack", "porosity","misalignment"],
            value="crack",
            label="Select Defect"
        )
    ],
    outputs=gr.Image(type="pil"),
    title="AI Manufacturing Defect Generator",
    description="Generate realistic defect images using AI"
)

demo.launch()
```
